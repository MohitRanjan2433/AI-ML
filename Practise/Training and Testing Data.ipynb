```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('carprices2.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Car Mileage Vs Sell Price

```python
plt.scatter(df['Mileage'],df['Sell Price($)'])
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')

plt.plot([min(df['Mileage']),max(df['Mileage'])],[40000,0],color='red')
plt.show() 
```

# Car Age Vs Sell Price

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
plt.xlabel('Age(yrs)')
plt.ylabel('Sell Price($)')
plt.title('Car Age Vs Sell Price')

# Add a straight line
plt.plot([min(df['Age(yrs)']), max(df['Age(yrs)'])], [50000, 0], color='red')

plt.show()
```

```python
X = df[['Mileage','Age(yrs)']]
```

```python
y = df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3) #test_size use for splitting the data in ratio 70-30
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
```
