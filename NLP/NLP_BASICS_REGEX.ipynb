Retrive Phone Number

```python
chat1 = 'customer: you ask lot of questions 1235678912, abA@xyz.com, 9998812345'
chat2 = 'customer: here it is: (123)-567-8912, abX_82@xyz.com'
chat3 = 'customer: yes, phone: 1235678912 email: abc@xyz.com'
```

```python
#To match 10 digit phone number....

pattern = '\d{10}'

matches = re.findall(pattern, chat1)
matches
```

```python
#To match the phone number included in the bracket

pattern = '\(\d{3}\)-\d{3}-\d{4}'

matches = re.findall(pattern, chat2)
matches
```

```python
#To match both the pattern in same....

pattern = '\d{10}|\(\d{3}\)-\d{3}-\d{4}'

matches = re.findall(pattern, chat3)
matches
```

Retrive email-id

```python
#To match the email-id of different patterns

pattern = '[a-z0-9A-Z_]*@[a-zA-Z]*\.[a-zA-Z]*'

matches = re.findall(pattern, chat2)
email = matches[0]
email
```

Regex For information Extraction----->

```python
text='''
Born	Elon Reeve Musk
June 28, 1971 (age 50)
Pretoria, Transvaal, Sout Africa
Citizenship
South Africa (1971–present)
Canada (1971–present)
United States (2002–present)
Education	University of Pennsylvania (BS, BA)
Title
Founder, CEO and Chief Engineer of SpaceX
CEO and product architect of Tesla, Inc.
Founder of The Boring Company and X.com (now part of PayPal)
Co-founder of Neuralink, OpenAI, and Zip2
Spouse(s)
Justine Wilson
​
​(m. 2000; div. 2008)​
Talulah Riley
​
​(m. 2010; div. 2012)​
​
​(m. 2013; div. 2016)
'''
```

```python
def get_pattern_match(pattern, text):
  matches = re.findall(pattern, text)
  if matches:
    return matches[0]
```

```python
pattern = 'age (\d+)'

matches = re.findall(pattern, text)
matches
```

```python
get_pattern_match(pattern, text)
```

```python
get_pattern_match('Born(.*)', text)
```

```python
get_pattern_match('Born.*\n(.*)\(age', text)
```

```python
get_pattern_match('\(age.*\n(.*)', text)
```

Function to get full-information--->

```python
text='''
Born	William Henry Gates III
October 28, 1955 (age 67)
Seattle, Washington, U.S.
Education
Harvard University
(dropped out)
Occupations
Businessmaninvestorphilanthropistprogrammerwriter
Years active	1972–present
Known for	Co-founding of Microsoft and Bill & Melinda Gates Foundation
Title
Partial list of founded and chaired companies
Spouse	Melinda French
​
​(m. 1994; div. 2021)​
Children	3
Parents
Bill Gates Sr.
Mary Maxwell
Awards
Knight Commander of the Order of the British Empire (2005)
Padma Bhushan (2015)
Presidential Medal of Freedom (2016)
Hilal-e-Pakistan (2022)
Website	gatesnotes.com
'''
```

```python
def get_personal_details(text):
  age = get_pattern_match('age (\d+)', text)
  full_name = get_pattern_match('Born(.*)', text)
  birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
  birth_place = get_pattern_match('\(age.*\n(.*)', text)
  awards = get_pattern_match(r"Awards.*\n(.*)\n(.*)\n(.*)\n(.*)", text)

  return {
      'age' : int(age),
      'name': full_name.strip(),
      'birth_date' : birth_date.strip(),
      'birth_place' : birth_place.strip(),
      'Awards': awards
  }
```

```python
get_personal_details(text)
```

3. Companies in europe reports their financial numbers of semi annual basis and you can have a document like this. To exatract quarterly and semin annual period you can use a regex as shown below

Hint: you need to use (?:) here to match everything enclosed

```python
text = '''
Tesla's gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
BMW's gross cost of operating vehicles in FY2021 S1 was $8 billion.
'''
```

```python
pattern  = 'FY(\d{4} (?:Q[1-4]|S[1-2]))'

matches = re.findall(pattern, text)
matches
```
